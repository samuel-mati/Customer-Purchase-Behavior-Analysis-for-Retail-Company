```python
import pandas as pd
```

```python
df= pd.read_csv('datasets/customer_shopping_behavior.csv')
df.head()
```

```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(" ","_")
df.columns.tolist()
```

```python
df.head()
```

```python
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
df.columns
```

```python
# Create a column age group

labels= ['Young Adult', 'Adult', 'Middle Aged', 'Senior']

df['age_group']= pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Create purchase_frequency_days column

frequency_mapping={
    'Fortnightly':14, 
    'Weekly':7, 
    'Annually':365, 
    'Quarterly':90, 
    'Bi-Weekly':14,
    'Monthly':30, 
    'Every 3 Months':90
    
}
```

```python
df['frequency_of_purchases'].unique()
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)

df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
# Import necessary packages
from sqlalchemy import create_engine

# Define connection variables
username = 'root'       # e.g., 'root'
password = 'your_password'       # your MySQL password
host = 'localhost'               # or IP of MySQL server
port = 3306                      # default MySQL port
database = 'customer_behaviour'  # database name
table_name = 'customers'          # target table name

```

```python
import mysql.connector
from mysql.connector import errorcode
# Test connection and create database if it doesn't exist
try:
    # Connect to MySQL server (no database specified yet)
    conn = mysql.connector.connect(
        host=host,
        user=username,
        password=password,
        port=port
    )
    cursor = conn.cursor()

    # Create database if it doesn't exist
    cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
    print(f"Database '{database}' is ready.")

except mysql.connector.Error as err:
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
        print("Access denied: check your username or password")
    else:
        print(f"Error: {err}")

finally:
    cursor.close()
    conn.close()
```

```python
# Create SQLAlchemy engine
engine = create_engine(f'mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}')

# Upload dataframe to MySQL
df.to_sql(
    name=table_name,
    con=engine,
    if_exists='replace',   
    index=False,
    chunksize=1000        # optional, useful for large datasets
)

print(f"{len(df)} rows uploaded successfully to table '{table_name}' in database '{database}'.")
```

```python
pd.read_sql('SELECT * FROM customers LIMIT 5;', engine)
```
